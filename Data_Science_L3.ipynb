```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apple_weight = [0.895, 0.91, 0.923, 0.926, 0.935, 0.941]
years = [2017, 2018, 2019, 2020, 2021, 2022]
plt.plot(years, apple_weight)
plt.xlabel('Year')
plt.ylabel('Tons per hectare')
```

```python
apple_weight = [0.895, 0.91, 0.923, 0.926, 0.935, 0.941]
orange_weight = [0.986, 0.952, 0.933, 0.926, 0.915, 0.891]
years = [2017, 2018, 2019, 2020, 2021, 2022]
plt.plot(years, apple_weight)
plt.plot(years, orange_weight)
plt.xlabel('Year')
plt.ylabel('Tons per hectare')

plt.title('Oranges and Apples Yield Comparison')
plt.legend(['Apples', 'Oranges'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(years, apple_weight, marker='o')
plt.plot(years, orange_weight, marker='x')
plt.xlabel('Year')
plt.ylabel('Tons per hectare')

plt.title('Oranges and Apples Yield Comparison')
plt.legend(['Apples', 'Oranges'])
```

```python
apple_weight = [0.495, 0.91, 0.823, 0.926, 0.735, 0.541]
orange_weight = [0.386, 0.252, 0.833, 0.626, 0.402, 0.391]
plt.bar(years, orange_weight, bottom=orange_weight)
plt.bar(years, apple_weight)
plt.xlabel('Year')
plt.ylabel('Tons per hectare')
```

```python
flowers = sns.load_dataset('iris')
flowers
```

```python
plt.title('Distribution of Sepal Width')
plt.hist(flowers.sepal_width, bins=5)
```

```python
import numpy as np
plt.title('Distribution of Sepal Width')
plt.hist(flowers.sepal_width, bins=np.arange(2, 5, 0.25))
```

```python
import numpy as np
plt.title('Distribution of Sepal Width')
plt.hist(flowers.sepal_width, bins=[1, 3, 4, 4.5])
```

```python
flowers.species.unique()
```

```python
setosa = flowers[flowers.species == 'setosa']
virginica = flowers[flowers.species == 'virginica']
versicolor = flowers[flowers.species == 'versicolor']
plt.hist(setosa.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
plt.hist(versicolor.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
```

```python
plt.title('Distribution of Sepal Width')
plt.hist([setosa.sepal_width, versicolor.sepal_width, virginica.sepal_width], bins=np.arange(2, 5, 0.25), stacked=True)
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
```

```python
plt.plot(flowers.sepal_length, flowers.sepal_width)
```

```python
plt.figure(figsize=(12, 6))
plt.title('Sepal Distribution')
sns.scatterplot(x=flowers.sepal_length, y=flowers.sepal_width, hue=flowers.species, s=70)
```
